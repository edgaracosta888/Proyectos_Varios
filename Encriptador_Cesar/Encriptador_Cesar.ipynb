# ENCRIPTADOR CESAR MEJORADO

```python
import string
# ====================================================================================================================
# Lista que contiene todos los caracteres que se pueden encriptar, incluyendo letras mayúsculas y minúsculas, dígitos,
# signos de puntuación y espacios
# ====================================================================================================================
abecedario = list(string.ascii_letters)
digitos = list(string.digits)
puntuacion = list(string.punctuation)
espacio = list(string.whitespace)
abc = abecedario + digitos + puntuacion + espacio

# ====================================================================================================================
# Función para separar el texto introducido por el usuario en una lista de caracteres y obtener el número entero para
# la encriptación
# ====================================================================================================================
def spliter() -> list:
    x = list(input("Introduce un texto: "))
    n = input("Introduce un número entero: ")
    return x, n

# ====================================================================================================================
# Función para encriptar el texto utilizando el cifrado César mejorado
# ====================================================================================================================
def encriptacion():
    texto, n = spliter()
    lista_encriptada = []
    for letter in texto:
        y = abc.index(letter) + int(n) # Suma el número entero al índice del carácter en la lista abc
        if y >= len(abc):
            y = y % len(abc)
        lista_encriptada.append(abc[y]) # Agrega el carácter encriptado a la lista de caracteres encriptados
    return lista_encriptada

# ====================================================================================================================
# Función para convertir la lista de caracteres encriptados de nuevo a una cadena de texto
# ====================================================================================================================
def lista_a_texto(lista):
    lista = ''.join(lista)
    return lista

# ====================================================================================================================
# Función principal para ejecutar el cifrado César mejorado
# ====================================================================================================================
def Encriptado_Cesar_Mejorado():
    lista = encriptacion()
    return lista_a_texto(lista)
```

```python
Encriptado_Cesar_Mejorado()
```

# DESENCRIPTADOR CESAR MEJORADO

```python
# ====================================================================================================================
# Función para separar el texto introducido por el usuario en una lista de caracteres y obtener el número entero para
# la desencriptación
# ====================================================================================================================
def desencriptar() -> list: 
    x = list(input("Introduce el texto a desencriptar: "))
    n = input("Introduce la clave de desencriptado: ")
    return x, n

# ====================================================================================================================
# Función para desencriptar el texto utilizando el cifrado César mejorado
# ====================================================================================================================
def desencriptacion():
    texto, n = desencriptar()
    lista_desencriptada = []
    for letter in texto:
        y = abc.index(letter) - int(n)
        if y < -len(abc):
            y = y % len(abc)
        lista_desencriptada.append(abc[y])
    return lista_desencriptada

# ====================================================================================================================
# Función para convertir la lista de caracteres desencriptados de nuevo a una cadena de texto
# ====================================================================================================================
def lista_a_texto(lista):
    lista = ''.join(lista)
    return lista
# ====================================================================================================================
# Función principal para ejecutar la desencriptación utilizando el cifrado César mejorado
# ====================================================================================================================
def Desencriptado_Cesar_Mejorado():
    lista = desencriptacion()
    return lista_a_texto(lista)
```

```python
Desencriptado_Cesar_Mejorado()
```
