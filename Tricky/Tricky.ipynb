# TRICKY (Tres en Línea)

```python
Board = [[i + 1 for i in range(3)], [j for j in range(4, 7)], [k for k in range (7, 10)]]
jugadas_IA = []
jugadas_US = []
numeros = list(range(1,10))
numeros.remove(5)
JG1 = set([1,2,3])
JG2 = set([4,5,6])
JG3 = set([7,8,9])
JG4 = set([1,4,7])
JG5 = set([2,5,8])
JG6 = set([3,6,9])
JG7 = set([1,5,9])
JG8 = set([3,5,7])

# ============================================================================================
# Función para que la computadora elija un número al azar y actulice la lista números posiblles
# ============================================================================================
def azar(lista):
    # toma una lista, la ordena aleatoriamente y devuelve el prier nùmero de la lista aleatoriamente ordenada y la lista sin este elemento.
    import random
    random.shuffle(lista)      # Los desordena aleatoriamente
    pos = lista[0]
    lista = lista[1:]
    return pos, lista

# ============================================================================================
# Función para actualziar la lista de números posibles después de la jugada del usuario
# ============================================================================================
def actualizar(P, lista):
    lista.remove(P)
    return lista

# ============================================================================================
# Función para crear y actualizar el tablero
# ============================================================================================

def tablero(x, PJ = numeros, JIA = jugadas_IA, JU = jugadas_US):
    # pinta el tablero de TRICKY con la primer jugada definida por el argumento de la funciòn
    for w in range(3):
        for i in range(3):
            print ('+', 7*'-', sep = '', end ='')
        print('+')

        for j in range(4):
            print('|', 7*' ', sep = '', end = '')
        print('')

        for j in range(3):
            if Board[w][j] in JIA:
                    print('|', 3*' ', 'X', 3*' ', sep = '', end = '')
            elif Board[w][j] in JU:
                    print('|', 3*' ', 'O', 3*' ', sep = '', end = '')
            else:
                print('|', 3*' ', Board[w][j], 3*' ', sep = '', end = '')
        print('|')

        for j in range(4):
            print('|', 7*' ', sep = '', end = '')
        print('')
    for i in range(3):
        print ('+', 7*'-', sep = '', end ='')
    print('+')

# ============================================================================================
# Función para iniciar y finalizar el juego
# ============================================================================================
def Tricky(x = 5, PJ = numeros, JIA = jugadas_IA, JU = jugadas_US):
    # PJ: Posiibles jugadas
    # JIA: Jugadas hechas por la computadora
    # JU: Jugadas hechas por el usuario
    JIA.append(x)
    #print(JIA, JU, PJ)
    tablero(x)
    if JG1.issubset(JU) or JG2.issubset(JU) or JG3.issubset(JU) or JG4.issubset(JU) or JG5.issubset(JU) or JG6.issubset(JU) or JG7.issubset(JU) or JG8.issubset(JU):
         print('Has ganado')
    elif JG1.issubset(JIA) or JG2.issubset(JIA) or JG3.issubset(JIA) or JG4.issubset(JIA) or JG5.issubset(JIA) or JG6.issubset(JIA) or JG7.issubset(JIA) or JG8.issubset(JIA):
         print('La máquina gana')
    elif len(PJ) == 0:
         print('Empate')
    else:
        print('Ingresa tu movimiento:')
        P = int(input('Ingresa tu movimiento: '))
        JU.append(P)
        actualizar(P, PJ)
        IA, PJ = azar(PJ)
        JIA.append(IA)
        #print(JIA, JU, PJ)
        Tricky(IA, PJ, JIA, JU)
        
          
```
