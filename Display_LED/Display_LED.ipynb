# LED DISPLAY V 1.0.
Usa un. diccionario y compara cada dígito del número ingresado con las llaves del diccionario para imprimir su correspondiente valor.

```python
# =========================================================================================================
# Función que recibe un número entero n y lo muestra en formato LED utilizando caracteres del diccionario
# =========================================================================================================
def led_v1():
# =========================================================================================================
# Diccionario que representa cada dígito del 0 al 9 en formato LED
# =========================================================================================================
    dic = {
    '1': ['  #', '  #', '  #', '  #', '  #'],
    '2': ['###', '  #', '###', '#', '###'],
    '3': ['###', '  #', '###', '  #', '###'],
    '4': ['# #', '# #', '###', '  #', '  #'],
    '5': ['###', '#', '###', '  #', '###'],
    '6': ['###', '#', '###', '# #', '###'],
    '7': ['###', '  #', '  #', '  #', '  #'],
    '8': ['###', '# #', '###', '# #', '###'],
    '9': ['###', '# #', '###', '  #', '  #'],
    '0': ['###', '# #', '# #', '# #', '###']
    }
# =========================================================================================================
# Transformar el número n en una lista de caracteres para poder acceder a cada dígito individualmente
# =========================================================================================================
    n = input("Introduce un número entero: ")
    if n.isdigit():

        n = ' '.join(n).split()
    # =========================================================================================================
    # Imprimir cada fila de los dígitos del diccionario correspondientes al número n, asegurando que estén
    # alineados correctamente
    # =========================================================================================================
        for i in range(len(dic['1'])):
            if i > 0:
                print()  # Imprime un salto de línea después de cada fila, excepto antes de la primera
            for j in n:
                if len(dic[j][i]) == 3:
                    print(dic[j][i], end ='  ')
                elif len(dic[j][i]) == 2:
                    print(dic[j][i], end ='   ')
                elif len(dic[j][i]) == 1:
                    print(dic[j][i], end ='    ')
    else:
        print("Por favor, introduce un número entero válido.")
        led_v1()
```
