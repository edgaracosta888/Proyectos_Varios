```python
import string
abecedario = ','.join(string.ascii_letters).split(',')
digitos = ','.join(string.digits).split(',')
puntuacion = ','.join(string.punctuation).split(',')
espacio = ','.join(string.whitespace).split(',')
abc = abecedario + digitos + puntuacion + espacio

def spliter() -> list: 
    x = input("Introduce un texto: ")
    x = ','.join(x).split(',')
    n = input("Introduce un número entero: ")
    return x, n

def encriptacion():
    texto, n = spliter()
    lista_encriptada = []
    for letter in texto:
        y = abc.index(letter) + int(n)
        if y >= len(abc):
            y = y % len(abc)
        lista_encriptada.append(abc[y])
    return lista_encriptada

def lista_a_texto(lista):
    lista = ''.join(lista)
    return lista

def Encriptado_Cesar_Mejorado():
    lista = encriptacion()
    return lista_a_texto(lista)
```

```python
Encriptado_Cesar_Mejorado()
```

```python
def desencriptar() -> list: 
    x = input("Introduce el texto a desencriptar: ")
    x = ','.join(x).split(',')
    n = input("Introduce la clave de desencriptado: ")
    return x, n


def desencriptacion():
    texto, n = desencriptar()
    lista_desencriptada = []
    for letter in texto:
        y = abc.index(letter) - int(n)
        if y < -len(abc):
            y = y % len(abc)
        lista_desencriptada.append(abc[y])
    return lista_desencriptada

def lista_a_texto(lista):
    lista = ''.join(lista)
    return lista

def Desencriptado_Cesar_Mejorado():
    lista = desencriptacion()
    return lista_a_texto(lista)
```
